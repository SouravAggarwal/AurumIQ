```python
import pandas as pd

# Define ETF mapping with only Index and ETF Name
etf_mapping = [
    {'Symbol': 'NIFTY INDIA DEFENCE', 'ETF Name': 'Groww Nifty India Defence ETF FoF – Direct Plan', "BuyPlan": "Groww Nifty India Defence ETF FoF, Motilal Oswal Nifty India Defence Index Fund"}, # Groww Nifty India Defence ETF MF
    {'Symbol': 'Nifty Capital Markets', 'ETF Name': 'Motilal Oswal Nifty Capital Market ETF'},
    {'Symbol': 'Nifty CPSE', 'ETF Name': 'CPSE ETF by Nippon India Mutual Fund'},
    {'Symbol': 'NIFTY MICROCAP250', 'ETF Name': 'Motilal Oswal Nifty Microcap 250 Index Fund'},
    {'Symbol': 'Nifty MidSmall Financial Services', 'ETF Name': 'Motilal Oswal Nifty MidSmall Financial Services Index Fund'},
    {'Symbol': 'Nifty200 Value 30', 'ETF Name': 'ICICI Prudential Nifty200 Value 30 ETF'},
    {'Symbol': 'NIFTY500 VALUE 50', 'ETF Name': 'Axis Nifty500 Value 50 ETF'},
    {'Symbol': 'Nifty Midcap 50', 'ETF Name': 'Kotak Nifty Midcap 50 ETF'},
    {'Symbol': 'Nifty PSE', 'ETF Name': 'Aditya Birla Sun Life Nifty PSE ETF'}
]

etf_df = pd.DataFrame(etf_mapping)
```

```python
import pandas as pd
import os

# Base path
base_path = "/Users/souravaggarwal/Downloads"

# List to hold individual dataframes
df_list = []

# Loop through the 5 files
for i in range(1, 5):
    file_name = f"Returnprofile_Data ({i}).csv"
    file_path = os.path.join(base_path, file_name)

    # Read the CSV file
    df = pd.read_csv(file_path, sep=',')  # tab-separated file as per your sample
    df['file_number'] = i  # Add file number column
    df_list.append(df)

# Combine all into one DataFrame
df = pd.concat(df_list, ignore_index=True)

# Display the final DataFrame
df
```

```python
df['1M'] = pd.to_numeric(df['1M'], errors='coerce')
df['3M'] = pd.to_numeric(df['3M'], errors='coerce')
df['1Yr'] = pd.to_numeric(df['1Yr'], errors='coerce')
df['3Yr'] = pd.to_numeric(df['3Yr'], errors='coerce')
df['5Yr'] = pd.to_numeric(df['5Yr'], errors='coerce')
df['10Yr'] = pd.to_numeric(df['10Yr'], errors='coerce')


# Add average column
df['avg_1_3_Yr'] = df[['1Yr', '3Yr']].mean(axis=1)
df['avg_1_3_5_Yr'] = df[['1Yr', '3Yr', '5Yr']].mean(axis=1)
```

```python

# Filter
filtered_df = df[df['avg_1_3_Yr'].notna() & (df['avg_1_3_Yr'] > 12)]
filtered_df2 = df[df['avg_1_3_5_Yr'].notna() & (df['avg_1_3_5_Yr'] > 12)]
filtered_1_3_5_Yr_positive = df[df['avg_1_3_5_Yr'].notna() & (df['avg_1_3_5_Yr'] > 12) & df['avg_1_3_Yr'].notna() & (df['avg_1_3_Yr'] > 12)]

sorted_df = filtered_1_3_5_Yr_positive.sort_values(by='avg_1_3_5_Yr', ascending=False)
sorted_df
```

```python
# Merge the ETF information with your existing DataFrame
merged_etf_df = sorted_df.merge(etf_df, on='Symbol', how='left')
merged_etf_df
```
